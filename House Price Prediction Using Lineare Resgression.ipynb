### select feature and target

```python
# independent feature
feature=['GrLivArea']

# dependent feature
target='SalePrice'
```

```python
x=data[feature]
x
```

```python
y=data[target]
y
```

### split the data into training and testing dataset

```python
from sklearn.model_selection import train_test_split_test_split
```

```python
#split data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2
                                               ,random_state=42)
```

### Train the linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
# create and train the model
model=LinearRegression()
model.fit(x_train,y_train)
```

### make prediction

```python
prediction=model.predict(x_test)
```

### Evaluate the model

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,prediction)
```

```python
prediction
```

```python
print(f"Mean squared Error:{mse}")
```

### visualize the results

```python
plt.scatter(x_test,y_test,color='black',label='Actual Price')
plt.scatter(x_test,prediction,color='red',label='Predicted Price')
plt.xlabel('Living Area (GrLiveArea)')
plt.ylabel('sal Price')
plt.legend()
plt.show()
```
